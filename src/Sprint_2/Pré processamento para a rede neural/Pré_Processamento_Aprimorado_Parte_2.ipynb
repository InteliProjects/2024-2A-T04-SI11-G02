**FUNÇÃO**

Está função extrai o ano e o mês de uma coluna de datas do DataFrame, criando uma nova coluna chamada ANOMES no formato YYYYMM.

```python
def extrairAnomes(df, col):

    df[col] = pd.to_datetime(df[col], format='%Y-%m-%d')

    df['ANOMES'] = df[col].dt.strftime('%Y%m')

    return df
```

### 3.1.1) Contabilizar Frequência de Fraudes

A função abaixo, visa contabilizar a frequência de fraudes que cada matrícula possui.

**ENTRADA**

```python
# BASE_FRAUDES
```

**FUNÇÃO**

Esta função contabiliza quantas vezes cada valor aparece na coluna de matrícula (col_matricula) do DataFrame (df) e armazena essa contagem em uma nova coluna chamada CONTAGEM_MATRICULA.

```python
def contabilizarFrequencia(df, col_matricula):

    df['CONTAGEM_MATRICULA'] = df.groupby(col_matricula)[col_matricula].transform('count')

    return df
```

**FUNÇÃO**

Está função une três DataFrames (df1, df2, df3) com base em uma coluna de chave comum, como a coluna de matrícula (col_matricula).

```python
def unirBases(df1, df2, col_matricula):
    # Unir a base 1 e 2
    df_merged = pd.merge(df1, df2, on=col_matricula, how='outer')
    return df_merged
```

### Normalização e one hot encoding das colunas de matrícula

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

# Supondo que 'BASE_FINAL' seja o DataFrame inicial
df = BASE_FINAL.copy()

# Converter a coluna 'ANOMES' para datetime, assumindo que o formato é 'AAAAMM'
df['ANOMES'] = pd.to_datetime(df['ANOMES'], format='%Y%m')

# Ordenar por 'ANOMES' para garantir a ordem correta
df = df.sort_values(['MATRICULA', 'ANOMES'])

# Filtrar os registros para remover aqueles onde 'CONS_MEDIDO' é NaN ou 0
df = df[df['CONS_MEDIDO'].notna()]

# Selecionar os últimos 12 registros por matrícula, independente do ano e da ordem dos meses
df['RANK'] = df.groupby('MATRICULA')['ANOMES'].rank(method='first', ascending=False)

# Filtrar para manter apenas as matrículas que possuem exatamente 12 registros
df = df[df['RANK'] <= 36]
df = df.groupby('MATRICULA').filter(lambda x: len(x) == 36)

# Criar um identificador de ordem dos últimos 12 meses
df['MES'] = df.groupby('MATRICULA').cumcount() + 1

# Pivotar os dados para ter 12 colunas, uma para cada mês (MES_1, MES_2, ..., MES_12)
df_pivot = df.pivot_table(index='MATRICULA', columns='MES', values='CONS_MEDIDO')

# Renomear as colunas para MES_1, MES_2, ..., MES_12
df_pivot.columns = [f'MES_{int(col)}' for col in df_pivot.columns]

# Manter as outras colunas originais como estão (latitude, longitude, economias, etc.)
df_meta = df[['MATRICULA', 'CONS_MEDIDO', 'ANOMES', 'COD_LATITUDE', 'COD_LONGITUDE',
       'ECO_INDUSTRIAL', 'ECO_COMERCIAL', 'ECO_PUBLICA', 'ECO_OUTRAS',
       'CATEGORIA', 'COD_GRUPO', 'COD_SETOR_COMERCIAL', 'NUM_QUADRA',
       'COD_ROTA_LEITURA', 'FRAUDADOR', 'CONTAGEM_MATRICULA',
       'CLUSTER', 'QUADRANT_1', 'QUADRANT_2', 'QUADRANT_3', 'QUADRANT_4']].drop_duplicates('MATRICULA')

# Merge com os dados pivotados
df_final = df_meta.merge(df_pivot, on='MATRICULA', how='left')

# Criar colunas para cada categoria
categorias = pd.get_dummies(df_final['CATEGORIA'], prefix='CATEGORIA').astype(int)

# Adicionar as novas colunas de categorias ao DataFrame final
df_final = pd.concat([df_final, categorias], axis=1)

# Remover a coluna 'CATEGORIA' original, pois agora temos as colunas individuais
df_final = df_final.drop(columns=['CATEGORIA'])

# Identificar as colunas a serem normalizadas (somente numéricas)
columns_to_normalize = df_final.select_dtypes(include=['int64', 'float64']).columns

# Normalizar as colunas usando RobustScaler
scaler = RobustScaler()
df_final[columns_to_normalize] = scaler.fit_transform(df_final[columns_to_normalize])

# Exibir o DataFrame final
df_final.head()
```
